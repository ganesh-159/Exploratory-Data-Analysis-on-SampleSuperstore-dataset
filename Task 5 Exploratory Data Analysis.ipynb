# Exploratory Data Analysis on SampleSuperstore dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('SampleSuperstore.csv')
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['Ship Mode'].value_counts()
```

```python
data.plot(kind='scatter',x='Sales',y='Profit');
data.plot()
```

```python
plt.figure(figsize=(15,7))
plt.bar('Sub-Category','Category',data=data)
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(data,hue='Ship Mode',size=5).map(plt.scatter,'Sales','Profit').add_legend();
plt.show()
```

```python
sns.FacetGrid(data,hue='Ship Mode',height=5).map(sns.distplot,'Sales').add_legend();
plt.show()
```

```python
sns.FacetGrid(data,hue='Ship Mode',height=5).map(sns.distplot,'Profit').add_legend();
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(data,hue='Region',size=5).map(plt.scatter,'Sales','Profit').add_legend();
plt.show()
```

```python
sns.FacetGrid(data,hue='Region',height=5).map(sns.distplot,'Sales').add_legend();
plt.show()
```

```python
sns.FacetGrid(data,hue='Region',height=5).map(sns.distplot,'Profit').add_legend();
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x='Sub-Category',hue='Region',data=data)
plt.show()
```

```python
plt.figure(figsize=(12,10))
data['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
```

```python
data.plot(kind='scatter',x='Sales',y='Quantity');
data.plot()
```

```python
data.plot(kind='scatter',x='Sales',y='Discount');
data.plot()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(data,hue='Sub-Category',size=5).map(plt.scatter,'Sales','Profit').add_legend();
plt.show()
```
